```python
import pandas as pd
```

```python
cluster_data = pd.read_csv(
    "https://raw.githubusercontent.com/MoH-Malaysia/covid19-public/main/epidemic/clusters.csv"
)
data = pd.DataFrame(cluster_data)
cluster = data.reset_index(inplace=True)
cluster = data.rename(columns={"index": "cluster_id"})
```

```python
# Word boundry \b
# ref: https://stackoverflow.com/questions/11350770/filter-pandas-dataframe-by-substring-criteria Ans by: Philipp Schwarz
cluster.loc[cluster["state"].astype(str).str.contains(r'\b14\b')].to_dict('records')
# cluster['state'].str.split(',', expand=True)
```

```python

cluster.iloc[1].to_dict()  # return single data
```

```python
last_update = cluster.iloc[-1:]["date_announced"].item()
cluster_list_data = cluster.query(f"date_announced == '{last_update}' and status == 'active'").sort_values('cases_total', ascending=False)
```
